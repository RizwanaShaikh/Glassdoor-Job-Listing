## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

df = pd.read_csv("C:/Users/bhavy/OneDrive/Desktop/glassdoor_jobs (1).csv")

df["Salary Estimate"] = df["Salary Estimate"].str.replace(r" \(Glassdoor est.\)", "", regex=True)
df["Min Salary"] = df["Salary Estimate"].str.extract(r'(\d+)K')[0].astype(float) * 1000
df["Max Salary"] = df["Salary Estimate"].str.extract(r'-(\d+)K')[0].astype(float) * 1000
df["Average Salary"] = (df["Min Salary"] + df["Max Salary"]) / 2
df["Company Name"] = df["Company Name"].str.split("\n").str[0]
df["Competitors"] = df["Competitors"].replace("-1", pd.NA)
df["Job State"] = df["Location"].apply(lambda x: x.split(",")[-1].strip())
df["Company Age"] = df["Founded"].apply(lambda x: 2024 - x if x > 0 else None)

import pandas as pd
import numpy as np
import re

file_path = "C:/Users/bhavy/OneDrive/Desktop/glassdoor_jobs (1).csv"
df = pd.read_csv(file_path)

df.replace(-1, np.nan, inplace=True)

df.drop(columns=["Unnamed: 0"], inplace=True, errors='ignore')

def clean_salary(salary):
    """
    Cleans and extracts numerical salary values.
    """
    salary = str(salary).lower()
    if "per hour" in salary or "employer provided salary" in salary:
        return np.nan, np.nan  # Ignore hourly rates & employer-provided estimates
    
    salary = re.sub(r"[^\d-]", "", salary)  # Remove non-numeric characters except '-'
    salary_range = salary.split("-")
    
    if len(salary_range) == 2:
        try:
            min_salary = int(salary_range[0])
            max_salary = int(salary_range[1])
            return min_salary, max_salary
        except ValueError:
            return np.nan, np.nan  # Handle cases where parsing fails
    
    return np.nan, np.nan

if "Salary Estimate" in df.columns:
    df["Min Salary"], df["Max Salary"] = zip(*df["Salary Estimate"].apply(clean_salary))
    df.drop(columns=["Salary Estimate"], inplace=True)  # Dropping original column

df.columns = df.columns.str.lower().str.replace(" ", "_")

df.drop_duplicates(inplace=True)

if "founded" in df.columns:
    df["founded"] = df["founded"].astype("Int64")  # Ensuring integer type
if "rating" in df.columns:
    df["rating"] = df["rating"].astype("float")  # Ensuring float type

numerical_cols = ["rating", "founded", "min_salary", "max_salary"]
df.describe(include="all")
````

# Removed Duplicates: Checked and dropped any duplicate rows.

# Ensured Correct Data Types: Converted "founded" to integer type and "rating" to float.

````python
# Chart - 1 Top job titles with a 5-star rating

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
file_path = "C:/Users/bhavy/OneDrive/Desktop/glassdoor_jobs (1).csv"  # Adjust path
df = pd.read_csv(file_path)

# Print available columns to verify correct naming
print("Columns in dataset:", df.columns)

# Ensure correct column names
rating_col = "Rating"  # Change if different in your dataset
job_col = "Job Title"
company_col = "Company Name"

# Handle missing values (if any)
df = df[[job_col, company_col, rating_col]].dropna()

# Filter top-rated jobs (rating = 5)
df_top_rated = df[df[rating_col] == 5]

# Get top 10 highest-rated jobs
top_rated_jobs = df_top_rated.groupby([job_col, company_col])[rating_col].count().reset_index()
top_rated_jobs = top_rated_jobs.sort_values(by=rating_col, ascending=False).head(10)

# Set Seaborn style
sns.set_style("whitegrid")

# Create bar plot
plt.figure(figsize=(12, 6))
sns.barplot(
    x=top_rated_jobs[rating_col], 
    y=top_rated_jobs[job_col] + " @ " + top_rated_jobs[company_col], 
    hue=top_rated_jobs[job_col], 
    legend=False, 
    palette="viridis"
)

# Labels and title
plt.xlabel("Number of Top Ratings (5-Star)")
plt.ylabel("Job Title @ Company")
plt.title("Top 10 Highest-Rated Job Titles & Companies")

# Show chart
plt.show()
````

#### Chart - 7

````python
# Chart - 7 Revenue by Sector Across Locations

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "C:/Users/bhavy/OneDrive/Desktop/glassdoor_jobs (1).csv" 
df = pd.read_csv(file_path)

if {"Sector", "Revenue", "Location"}.issubset(df.columns):
    df_filtered = df.dropna(subset=["Sector", "Revenue", "Location"])
    
    df_filtered["Revenue"] = df_filtered["Revenue"].astype(str).str.extract(r"(\d+)").astype(float)
        
    df_filtered = df_filtered[df_filtered["Revenue"] > 0]

    sector_revenue = df_filtered.groupby(["Sector", "Location"])["Revenue"].sum().reset_index()
    
    top_sectors = sector_revenue.groupby("Sector")["Revenue"].sum().nlargest(10).reset_index()
    
    top_sectors_df = sector_revenue[sector_revenue["Sector"].isin(top_sectors["Sector"])]

    if not top_sectors_df.empty:
        pivot_df = top_sectors_df.pivot(index="Sector", columns="Location", values="Revenue").fillna(0)
        
        plt.figure(figsize=(14, 7))
        pivot_df.plot(kind="bar", stacked=True, colormap="viridis", figsize=(14, 7))
        
        plt.xlabel("Sector")
        plt.ylabel("Total Revenue ($M)")
        plt.title("Top 10 Sectors by Revenue across Locations")
        plt.xticks(rotation=45, ha="right")
        plt.legend(title="Location", bbox_to_anchor=(1.05, 1), loc="upper left")

        plt.show()
    else:
        print("No valid data available for visualization.")

else:
    print("Required columns not found in dataset.")
````

````python
# Chart - 8 Highest Job Titles vs. Revenue & Location

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "C:/Users/bhavy/OneDrive/Desktop/glassdoor_jobs (1).csv" 
df = pd.read_csv(file_path)

if {"Job Title", "Revenue", "Location"}.issubset(df.columns):

    df_filtered = df.dropna(subset=["Job Title", "Revenue", "Location"])
    
    df_filtered["Revenue"] = df_filtered["Revenue"].astype(str).str.extract(r"(\d+)").astype(float)
    
    df_filtered = df_filtered[df_filtered["Revenue"] > 0]

    job_revenue = df_filtered.groupby(["Job Title", "Location"])["Revenue"].sum().reset_index()
    
    top_jobs = job_revenue.groupby("Job Title")["Revenue"].sum().nlargest(10).reset_index()
    
    top_jobs_df = job_revenue[job_revenue["Job Title"].isin(top_jobs["Job Title"])]

    if not top_jobs_df.empty:
        pivot_df = top_jobs_df.pivot(index="Job Title", columns="Location", values="Revenue").fillna(0)
        
        pivot_df.plot(kind="bar", colormap="coolwarm", figsize=(14, 7))
        
        plt.xlabel("Job Title")
        plt.ylabel("Total Revenue ($M)")
        plt.title("Top 10 Job Titles by Revenue across Locations")
        plt.xticks(rotation=45, ha="right")
        plt.legend(title="Location", bbox_to_anchor=(1.05, 1), loc="upper left")

        plt.show()
    else:
        print("No valid data available for visualization.")

else:
    print("Required columns not found in dataset.")
````
